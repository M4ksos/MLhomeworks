### B. Пропуски и очистка

```python
miss_val = df.isnull().sum()
print(miss_val)
```

##### 1. Численные данные: заполняем медианой ( = устойчива к выбросам )
##### Если распределение нормальное, можно средним.

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].median())
print(df['Amount'])
```

##### 2. Категориальные данные: заполняем модой (самым частым значением)

```python
if 'Card_Type' in df.columns:
    card_mode = df['Carf_Type'].mode()[0]
    df['Card_Type'] = df['Card_Type'].fillna(card_mode)
else:
    print("bla-bla-bla")
```

```python
dfdrop = df.copy()
dfdrop = df.dropna()
print(dfdrop)
```

### C. Расширенная статистика

```python
print(f"Наименьшее значение:\t\t\t\t|  {df['Amount'].min()}")
print(f"Наибольшее значение:\t\t\t\t|  {df['Amount'].max()}")
print(f"Среднее значение столбца:\t\t\t|  {df['Amount'].mean()}")
print(f"Суммированное значение:\t\t\t\t|  {df['Amount'].sum()}")
print(f"Среднее значение в столбце:\t\t\t|  {df['Amount'].median()}")
print(f"Количество ненулевых значений в столбце:\t|  {df['Amount'].count()}")
print(f"Разброс значений в столбцк:\t\t\t|  {df['Amount'].std()}")
print(f"Изменчивость значений в столбце:\t\t|  {df['Amount'].var()}")
print(f"Ассиметрия (Skewness):\t\t\t\t|  {df['Amount'].skew():.2f}")
print(f"Эксцесс (Kurtosis):\t\t\t\t|  {df['Amount'].kurt():.2f}")
```

```python
print(df['Amount'].quantile([0.05,0.25,0.50,0.75,0.95]))
```

### D. Фичи: Энкодинг и инжиниринг

#### Категорий в датасете нет, поэтому OneHot делать не на чем, -> делаем Time Engineering

```python
df['Hour'] = (df['Time']// 3600) % 24
print(df['Hour'])

df['Log_Amount'] = np.log1p(df['Amount'])
print(df['Log_Amount'])
# пример группировки Future Engineering :

df['Is_HighValue'] = (df['Amount'] > df['Amount'].quantile(0.95)).astype(int)
print(df['Is_HighValue'])
```

### E. Визуализация

Seaborn: Распределение сумм

```python
plt.figure(figsize=(10,5))
sns.kdeplot(df[df['Class'] == 0]['Log_Amount'], label = 'Normal', fill = True)
sns.kdeplot(df[df['Class'] == 1]['Log_Amount'], label = 'Fraud', fill = True)
plt.title('Распределение логарифма суммы')
plt.legend()
plt.show()
```

Plotly: Интерактивный график зависимости

```python
fig = px.scatter(df.sample(5000), x="V1",y="V2",color="Class",title="Срез PCA feat V1 и V2",hover_data=['Amount','Hour'])
fig.show()
```

Матрица корреляций

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title('Матрица корреляций')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Class', y='Amount', data=df)
plt.yscale('log')
plt.title("Разброс сумм для разных классов")
plt.show()
```
